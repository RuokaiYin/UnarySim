```python
import torch
from UnarySim.sw.metric.metric import NormStability, Stability
from UnarySim.sw.stream.gen import RNG, SourceGen, BSGen
import random
```

```python
width = 8
length = 2**width
device = torch.device("cpu")


for ite in range(length+1):
    dim = 1
    val = 1.0*ite/length
    val_tensor = torch.tensor([val])
    val_bin = SourceGen(val_tensor, width, "unipolar")
    rng = RNG(width, dim, "Sobol").to(device)
    bs = BSGen(val_bin(), rng())
    normstb = NormStability(val_tensor, mode="unipolar", threshold=0.05).to(device)
    stb = Stability(val_tensor, mode="unipolar", threshold=0.05).to(device)
    for _ in range(length):
        a = bs(torch.tensor([_]))
        normstb.Monitor(a)
        stb.Monitor(a)
    print(ite, "val:", val, "final stability:", normstb().item())
#     result = normstb()
# #     if result >= 1:
#     print("iter:", ite, "\n", "val:", val, "\n", "stab:", stb(), "\n", "norm stab:", result, "\n", "max_stab:", normstb.max_stab, "\n", "R:", normstb.R, "\n", "l_p:", normstb.l_p)
#     print("stab", stb())
#     print("norm stab", normstb())
#     print("norm stab: max stab", normstb.max_stab)
```

```python
width = 6
length = 2**width

for ite in range(length+1):
    dim = 1
    val = 1.0*ite/length
    val_tensor = torch.tensor([[val, 1-val]]).cuda()
    val_bin = SourceGen(val_tensor, width, "unipolar")
    rng = RNG(width, dim, "Sobol").cuda()
    bs = BSGen(val_bin(), rng())
    normstb = NormStability(val_tensor, mode="unipolar", threshold=0.05).cuda()
    stb = Stability(val_tensor, mode="unipolar", threshold=0.05).cuda()
    for _ in range(length):
        a = bs(_)
        normstb.Monitor(a)
        stb.Monitor(a)
    print(ite, normstb(), normstb.max_stab_R*normstb.max_stab_l_p)
```
